```python
import pandas as pd
from collections import defaultdict
first=open("RESTRICTED_yoavf_4_18_2019_10_37_3.csv")
second=open("Table.Correlation.csv")
firstreader=csv.reader(first)
secondreader=csv.reader(second)
i=0
firstobject={}
for first in firstreader:
    if i==0:
        i+=1
        continue
    firstobject[int(first[0])]=first
family_ids={}
for second in secondreader:
    key=int(second[0].replace("'",""))
    if key in firstobject:
        family_ids[firstobject[key][5]]=True
groups=defaultdict(list)
for test in firstobject:
    if firstobject[test][5] in family_ids:
        groups[str(firstobject[test][5])].append(test)
print(groups)
```
